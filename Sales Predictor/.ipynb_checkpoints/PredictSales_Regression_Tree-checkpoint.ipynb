- Groupby the train data by month, item and shop as the test prediction is done MoM(Month-on-Month)
- This is done because the number of items need to be predicted month wise.

```python
# Mean is used as opposed to Median, because Mean captures the sudden increase of decrease in prices 
# which could have happened month based, like End of Season Sale etc.
month_wise = train.groupby(by=['item_id', 'shop_id', 'date_block_num']).agg({'item_price':np.mean, 
                                                                              'item_cnt_day': sum}).reset_index()
month_wise["id"] = month_wise["shop_id"].astype(str)+"_"+month_wise["item_id"].astype(str)
month_wise
```

```python
month_wise.id.value_counts()
```

```python
sample_entry = month_wise[month_wise.id == '25_16167']
sns.scatterplot(sample_entry.date_block_num, sample_entry.item_cnt_day)
```

```python
my_order = (1, 1, 1)
print(sample_entry.item_cnt_day)
ts_model = SARIMAX(sample_entry.item_cnt_day, order=my_order).fit()
ts_model.forecast()
```
